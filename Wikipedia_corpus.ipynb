## Import the libraries

```python
import wikipedia
from datasets import load_dataset
```

```python
# Check the summary method
wikipedia.summary("Computer Science")
```

```python
# Get a list of computer sciency topics
topics = wikipedia.page('Computer Science').links
topics
```

```python
# List of all computer science-related articles
len(topics)
```

```python
# Filter out irrelevant articles and add BST
removals = ['Computer science (disambiguation)', 'Computer science education', 'Biology', 'Oxford University Press',
            'New York City', 'National Curriculum', 'Natural science', 'Multi-disciplinary', 'Greek language',
            'Glossary of virology','Glossary of agriculture', 'Glossary of biology', 'Philosophy', 'United States']

result = []
result.append('Binary search tree')
for word in topics:
    if word not in removals:
        result.append(word)
```

```python
# List after filtering
len(result)
```

```python
# Create a corpus of articles
corpus = []
for key in result:
  corpus.append(wikipedia.page(key, auto_suggest=False).content)
```

```python
type(corpus)
```

```python
# Inspect the corpus
corpus[1]
```

```python
len(corpus)
```

```python
# Open the file, write mode, and duplicate the corpus
with open(r'test.txt', 'w') as fp:
    for item in corpus:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('txt file created')
```
